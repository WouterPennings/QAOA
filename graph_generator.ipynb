```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
LOW_CONNECTIVITY_FACTOR = 0.4
HIGH_CONNECTIVITY_FACTOR = 0.8
```

```python
def generate_connected_graph(n, p):
    """Generate a graph with no isolated nodes."""
    while True:
        graph = nx.gnp_random_graph(n, p)
        if nx.is_connected(graph):
            return graph
```

```python
graphs = []

for n in range(4, 11):
    graph_A = generate_connected_graph(n, LOW_CONNECTIVITY_FACTOR)
    graph_B = generate_connected_graph(n, HIGH_CONNECTIVITY_FACTOR)
    graphs.append([graph_A, graph_B])
```

```python
for i, (graph_A, graph_B) in enumerate(graphs):
    # Save Graph A
    plt.figure(figsize=(6, 6))
    nx.draw(graph_A, with_labels=True, node_color='skyblue', edgecolors="black", node_size=1000)
    plt.title(f'Graph with {graph_A.number_of_nodes()} nodes, low connectivity')

    plt.savefig(f'data_noiseless/graph_{len(graph_A.nodes)}_low.png', bbox_inches='tight')
    plt.show()
    plt.close()

    # Save Graph B
    plt.figure(figsize=(6, 6))
    nx.draw(graph_B, with_labels=True, node_color='skyblue', edgecolors="black", node_size=1000)
    plt.title(f'Graph with {graph_B.number_of_nodes()} nodes, high connectivity')

    plt.savefig(f'data_noiseless/graph_{len(graph_A.nodes)}_high.png', bbox_inches='tight')
    plt.show()
    plt.close()
```
